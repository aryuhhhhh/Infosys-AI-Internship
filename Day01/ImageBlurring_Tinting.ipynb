```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# Opens a image in RGB mode
im = Image.open(r"/content/image.avif")

# Blurring the image
im1 = im.filter(ImageFilter.BoxBlur(3))

# Shows the image in image viewer
plt.imshow(im1)
plt.title("Blurred Image")
plt.axis('off')
```

```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# Opens a image in RGB mode
im = Image.open(r"/content/image.avif")

# Blurring the image
im1 = im.filter(ImageFilter.BoxBlur(3))

# Shows the image in image viewer
plt.imshow(im1)
plt.title("Blurred Image")
plt.axis('on')
```

```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt

im = Image.open("/content/image.avif").convert("RGB")

blue_tint = Image.new("RGB", im.size, (255, 0, 255))

tinted = Image.blend(im, blue_tint, alpha=0.5)

plt.imshow(tinted)
plt.axis('off')
plt.title("Blue-Tinted Image")
plt.show()
```
